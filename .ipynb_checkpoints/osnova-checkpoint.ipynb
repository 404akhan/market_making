```python
class Sim:
    def __init__(self, execution_latency: float, md_latency: float, max_trade_size: float) -> None:
        self.execution_latency = execution_latency
        self.md_latency = md_latency
        self.max_trade_size = max_trade_size
        
        self.i_tr = 0
        self.i_lobs = 0
        self.cur_ns = 0
        
        self.unique_id = 1
        self.orders = {'buy': {}, 'sell': {}}
        self.last_md = {}


    def my_iter(self):
        md = {}
        
        if btc_trades.iloc[self.i_tr]['exchange_ts'] < btc_lobs.iloc[self.i_lobs][' exchange_ts']:
            data = btc_trades.iloc[self.i_tr]
            swapped_side = 'bid' if (data['aggro_side'] == 'ASK') else 'ask'
            md[swapped_side] = data['price']

            self.i_tr += 1
            md['time_passed'] = (data['exchange_ts'] - self.cur_ns) / (1000 * 1000)
            self.cur_ns = data['exchange_ts']
        else:
            data = btc_lobs.iloc[self.i_lobs]
            md['ask'] = data['btcusdt:Binance:LinearPerpetual_ask_price_0']
            md['bid'] = data['btcusdt:Binance:LinearPerpetual_bid_price_0']

            self.i_lobs += 1
            md['time_passed'] = (data[' exchange_ts'] - self.cur_ns) / (1000 * 1000)
            self.cur_ns = data[' exchange_ts']
        
        if 'ask' in md:
            self.last_md['ask'] = md['ask']
        if 'bid' in md:
            self.last_md['bid'] = md['bid']
        self.last_md['nanosec'] = self.cur_ns
        
        return md
        

    def tick(self):
        self.prepare_orders()
        executed = self.execute_orders()
        
        

    def prepare_orders(self):
        pass

    
    def execute_orders(self):
        executed = []
        md = my_iter()
        
        if 'ask' in md:
            for uid, (price, size) in list(self.orders['buy'].iteritems()):
                if price >= md['ask']:
                    executed.append(uid)
                    cancel_order('buy', uid)
        
        if 'bid' in md:
            for uid, (price, size) in list(self.orders['sell'].iteritems()):
                if price <= md['bid']:
                    executed.append(uid)
                    cancel_order('sell', uid)

                    
    def place_order(self, side, price, size):
        if size > self.max_trade_size or side not in self.orders:
            return -1
        
        self.orders[side][self.unique_id] = (price, size)
        
        self.unique_id += 1
        return self.unique_id - 1

    
    def cancel_order(self, side, unique_id):
        return self.orders[side].pop(unique_id, -1)
```

```python
def load_md_from_file(pref, trades, lobs, ti, li) -> list[MdUpdate]:
    # TODO: load actual md
    pref_ask_price = pref + 'ask_price_'
    pref_ask_size = pref + 'ask_vol_'
    pref_bid_price = pref + 'bid_price_'
    pref_bid_size = pref + 'bid_vol_'
    
    mds = []
    i_tr = 0
    
    print('Started parsing...')
    for i_lobs in range(lobs.shape[0]):
        d = lobs[i_lobs]
        asks = [(d[li[pref_ask_price + str(i)]], d[li[pref_ask_size + str(i)]]) for i in range(10)]
        bids = [(d[li[pref_bid_price + str(i)]], d[li[pref_bid_size + str(i)]]) for i in range(10)]
        book = OrderbookSnapshotUpdate(timestamp = d[li[' exchange_ts']], asks = asks, bids = bids)
        
        anon_trades = []        
        while i_tr < trades.shape[0] and trades[i_tr][ti['exchange_ts']] < lobs[i_lobs][li[' exchange_ts']]:
            d = trades[i_tr]
            at = AnonTrade(timestamp=d[ti['exchange_ts']], side=d[ti['aggro_side']], size=d[ti['size']], price=d[ti['price']])    
            anon_trades.append(at)

            i_tr += 1
            if i_tr % 10000 == 0:
                print('Trade progress: %d, %.2f%%' % (i_tr, i_tr / trades.shape[0] * 100))
        
        mds.append(MdUpdate(orderbook = book, trades = anon_trades))
        
        if i_lobs % 10000 == 0:
            print('Lobs progress: %d, %.2f%%' % (i_lobs, i_lobs / lobs.shape[0] * 100))
            
    return mds
                 

loaded_md = load_md_from_file('btcusdt:Binance:LinearPerpetual_', np_btc_trades, np_btc_lobs, btc_ti, btc_li)
```
